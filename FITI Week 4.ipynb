```python
import numpy as np

patients_records = np.array([
    [25, 0, 4, 60, 10, 0, 350],
    [60, 2, 5, 30, 48, 1, 4200],
    [40, 1, 2, 20,  2, 0, 150],
    [10, 3, 3, 15,  6, 0, 200],
    [72, 0, 5, 90, 24, 1, 1800],
    [33, 1, 1, 10,  1, 0, 80],
    [55, 2, 4, 45, 36, 0, 3500],
    [28, 0, 3, 50,  8, 0, 500],
    [65, 2, 5, 25, 72, 1, 9000],
    [18, 3, 2, 12,  3, 0, 120],
    [49, 1, 3, 35,  4, 0, 220],
    [80, 0, 4, 110, 30, 1, 2500],
])
```

```python
#1 Total number of patients in the patients dataset.
Total_num = patients_records.shape[0]
print(f"Total number: {Total_num} patients \n")

# 2. Average age of patients.
Average_age = np.round(np.mean(patients_records[:,0]), 1)
print(f"Average age: {Average_age} \n")

# 3. Average waiting time (minutes)
Average_wait = np.round(np.mean(patients_records[:,3]), 1)
print(f"Average wait: {Average_wait} minutes \n")

# 4. Average length of stay (hours)
Average_length = np.round(np.mean(patients_records[:,4]), 1)
print(f"Average length: {Average_length} hours \n")

# 5. Total treatment cost for all patients.
Total_treatment = np.sum(patients_records[:,6])
print(f"Total_treatment: ${Total_treatment} \n")
```

```python
days = np.array(["Mon","Tue","Wed","Thu","Fri","Sat","Sun"])
departments = np.array(["Emergency", "Outpatient", "Surgery", "Pediatrics"])

daily_visits = np.array([
    [30, 50, 10, 20],
    [35, 55, 12, 18],
    [40, 60, 15, 22],
    [38, 58, 14, 25],
    [45, 65, 18, 30],
    [50, 40, 20, 15],
    [55, 35, 22, 10],
])
```

```python
# 1. Total visits in the week (all departments combined).
Total_visits = np.sum(daily_visits)
print(f"Total visit: {Total_visits} visits \n")

# 2. Total visits per department for the week.
Total_visits_dept = np.sum(daily_visits, axis = 0)
print(f"Total_visits_dept: {Total_visits_dept} \n")

# 3. Which day had the highest total visits?
daily_total = np.sum(daily_visits, axis = 1)
highest_day_index = np.argmax(daily_total)
print(f"Highest day: {days[highest_day_index]} ({daily_total[highest_day_index]} visits) \n")

# 4. Which department had the highest visits on Saturday?
sat = daily_visits[5]
sat_dept_index = np.argmax(sat)
print(f"Top dept on Saturday: {departments[sat_dept_index]} ({sat[sat_dept_index]} visits) \n")
```

```python
# 1. Count how many patients are in each department (Emergency/Outpatient/Surgery/Pediatrics).
for d in range(4):
    count = np.sum(patients_records[:, 1] == d)
    print(f"{departments[d]} count: {count} \n")
```

```python
# 2.Compute average waiting time per department.
for d in range(4):
    avg_wait = np.round(np.mean(patients_records[patients_records[:, 1] == d, 3]), 1)
    print(f"{departments[d]} avg wait: {avg_wait} min \n")
```

```python
# 3. Compute average cost per department.
for d in range(4):
    avg_cost = np.mean(patients_records[patients_records[:, 1] == d, 6])
    print(f"{departments[d]} avg cost: ${avg_cost:.0f} \n")
```

```python
# 4. Which department has the highest average waiting time?
avg_waits = np.array([np.mean(patients_records[patients_records[:, 1] == d, 3]) for d in range(4)])
highest   = np.argmax(avg_waits)
print(f"Highest avg wait: {departments[highest]} ({avg_waits[highest]:.1f} min) \n")
```
