```python
import pandas as pd
import numpy as np
from numpy import nan
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Making a copy of the original dataset
foodhub = foodhub_original.copy()
foodhub
```

```python
# Standardising column names
foodhub.columns = foodhub.columns.str.replace("_", " ").str.title().str.strip()
foodhub.columns
```

```python
# Checking descriptive statistics of the dataset
foodhub.describe()
```

```python
# Checking descriptive statistics of Food Preparation Time column
foodhub[["Food Preparation Time"]].describe()
```

```python
# Checking for number of unique values(nunique)
foodhub.nunique() 
```

```python
# Using expression statement to check for categorical features
cat_features = [col for col in foodhub.columns if foodhub[col].nunique() < 10 and foodhub[col].nunique() > 1]
cat_features
```

```python
# Checking for categorical features
for col in cat_features:
    print(str(col) + ':' + str(foodhub[col].unique()))
    print("="*80)
```

```python
# Changing cat features data type to category 
foodhub[cat_features] = foodhub[cat_features].astype("category")
```

```python
# Dropping ID columns
foodhub.drop(columns = ["Order Id", "Customer Id"], inplace = True)
```

```python
# Checking dataset new information
foodhub.info()
```

```python
# Checking the new dimension of the dataset
foodhub.shape
```

```python
# Checking the skewness of our data
perfect_skew = []
approximate_skew = []
moderate_skew = []
high_skew = []

for col in foodhub.select_dtypes(include = ["int64", "float64"]):
    if foodhub[col].skew() == 0:
         perfect_skew.append(col)
    elif abs(foodhub[col].skew()) < 0.5:
        approximate_skew.append(col)
    elif abs(foodhub[col].skew()) >= 0.5 and abs(foodhub[col].skew()) < 1:
        moderate_skew.append(col)
    else:
        high_skew.append(col)

print(f"Perfect Skewness: {perfect_skew}")
print(f"Approximate Skewness: {approximate_skew}")
print(f"Moderate Skewness: {moderate_skew}")
print(f"High Skewness: {high_skew}")
```

```python
# Visualizing the distribution of all numerical features using hisplot
for col in foodhub.select_dtypes(include = ["int64", "float64"]):
    plt.figure(figsize = (8,5))
    sns.histplot( data = foodhub, x = col, kde = True)
    plt.show()
```

```python
# Automated imputation of foodhub based on data type
for col in foodhub.columns:
    if foodhub[col].dtype in ["int64", "float64"]:
        if abs(foodhub[col].skew()) > 0.5:
            foodhub[col] = foodhub[col].fillna(foodhub[col].median())
        else:
            foodhub[col] = foodhub[col].fillna(foodhub[col].median())
    elif foodhub[col].dtypes in ["category", "str", "bool"]:
        foodhub[col] = foodhub[col].fillna(foodhub[col].mode()[0] if not foodhub[col].mode().empty else "Missing")

# Showcasing foodhub data
foodhub.isnull().sum()
```

```python
# Filtering out how many orders are not rated
foodhub["Rating"].value_counts()
```

```python
# Changing data types 
foodhub["Cost Of The Order"] = foodhub["Cost Of The Order"].astype(float)
foodhub["Food Preparation Time"] = foodhub["Food Preparation Time"].astype(int)
foodhub["Delivery Time"] = foodhub["Delivery Time"].astype(int)
for col in foodhub.select_dtypes(include = "str"):
    foodhub[col] = foodhub[col].astype("category")
foodhub
```

```python
# Filtering out the top 5 restaurants in terms of the number of orders received
restaurant_orders = foodhub["Restaurant Name"].value_counts().head(5)
restaurant_orders
```

```python
# Filtering out the most popular cuisine on weekends
weekend_cuisine = foodhub[(foodhub["Day Of The Week"] == "Weekend")]
weekend_cuisine = weekend_cuisine["Cuisine Type"].mode()[0]
print(f" The most popular cuisine on weekends is: {weekend_cuisine}")
```

```python
# Filtering out the percentage of the orders that cost more than 20 dollars
order_percentage = (foodhub["Cost Of The Order"] > 20).mean()  * 100
print(f" The percentage of orders costing more than 20 dollars is: {order_percentage:.2f}")
```

```python
# Filtering out the mean order delivery time
mean_delivery_time = foodhub["Delivery Time"].mean()
print(f" The mean of delivery time is: {mean_delivery_time:.2f}")
```
