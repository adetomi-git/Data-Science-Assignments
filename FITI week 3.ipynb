# Write a python program that:
- sets the user account balance and pin(1234)
- display a simple ATM menu:
1. check balance
2. withdraw money
3. deposit money
4. ask user to input choice
- use conditional statement to:
1. print current balance
2. ask for a withdrawal amount, check to see it is valid(positive <= balance)
3. update balance
4. ask for a deposit amount (must be positive) and updated balance.
5. for any other input (print "Invalid option selected)

```python
balance = 5000
default_pin = "1234" 

attempt = 1
access_granted = False

while attempt <= 3:
    print(f"Attempt {attempt} of 3")
    users_pin = input("Enter your pin: ")    
    
    if users_pin == default_pin:
        access_granted = True
        print("Correct pin! Welcome.")
        break

    else:
        trials_remaining = 3 - attempt

        if trials_remaining == 0:
            print("Incorrect pin!")
            print("Attempts exceeded.")
            print("Card has been blocked.")
            print("Please visit the closest branch to you.")
            break

        else:
            print("Incorrect pin! Please try again.")
            attempt += 1


if not access_granted:
    print("Access denied.")

if access_granted:

    while True:
        print("1. Check Balance")
        print("2. Withdraw Money")
        print("3. Deposit Money")
        print("4. Exit")

        try:
            operation = input("Enter menu option: ")

            if operation == "1":
                print(f"Your current balance is: {balance: }")

            elif operation == "2":
                try:
                    amount = int(input("Enter amount: "))
                    if amount <= 0:
                        print("Amount must be greater than 0.")
                    elif amount > balance:
                        print("Insufficient funds.")
                    else:
                        balance = balance - amount
                        print("Withdrawal successful!")
                        print(f"Withdrew {amount: }. New balance: {balance: }")
                except ValueError:
                    print("Invalid amount. Please enter a number.")

            elif operation == "3":
                try:
                    amount = int(input("Enter your deposit: "))
                    if amount <= 0:
                        print("Deposit amount must be greater than 0.")
                    else:
                        balance = balance + amount
                        print("Deposit successful!")
                        print(f"Deposited {amount: }. New balance: {balance: }")
                except ValueError:
                    print("Invalid amount. Please enter a number.")

            elif operation == "4":
                print("Thank you for banking with us. Goodbye!")
                break

            else:
                print("Invalid option. Please choose 1 - 4.")

        except ValueError:
            print("Please enter a number (1 - 4).")
```
