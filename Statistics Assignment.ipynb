### Case Studies
- You find r = 0.92 between ice cream sales and sunburn cases. Should you conclude ice cream causes sunburn? Explain?
- In an observational study, antidepressant use correlates with higher suicide risk. Does this mean anti depressants cause suicide risk? What else could explain this?
- Your ML model predicts that users who watch more tutorial videos have higher retention. Can you confidently say tutorial videos cause better retention.
- How would you design study to test whether a new drug reduces blood pressure(causually)?

#### You find r = 0.92 between ice cream sales and sunburn cases. Should you conclude ice cream causes sunburn? Explain?
- No. There cannot be a conclusion, because correlation doesn't imply causation. The ice cream sales and sunburn cases are likely increasing because of another common factor, which might be the sun or an extremely hot weather condition. 

#### In an observational study, antidepressant use correlates with higher suicide risk. Does this mean anti depressants cause suicide risk? What else could explain this?
- No, it doesn't. People prescribed with antidepressants already have severe depression, which itself might lead to suicide. A trial would be needed to test causation.

#### Your ML model predicts that users who watch more tutorial videos have higher retention. Can you confidently say tutorial videos cause better retention.
No. Users who watched the tutorial videos might be driven by the motivation to learn and acquire more understanding.

#### How would you design study to test whether a new drug reduces blood pressure(causually)?
Conduct a Randomized Controlled Trial (RCT) on two groups. Administer the new drug to the treatment group and a placebo to the control group, while measuring their blood pressure before and after the trial, then compare the results of both groups.
