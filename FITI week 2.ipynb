```python
##### Scholarship qualifier
CGPA = float(input("Enter your CGPA: "))
attendance = int(input("Enter your attendance: "))
disciplinary = input("Enter disciplinary: ")
recommendation = input("Enter recommendation: ")


if CGPA < 2.5 or disciplinary == "yes":
    print("Disqualified")
elif CGPA >= 4.5 and attendance > 75 and recommendation == "yes":
    print("Qualified")
elif CGPA >= 2.5 and attendance > 75 and recommendation == "yes" and disciplinary == "no":
    print("You are on awaiting list")
else:
    print("You're not eligible")
```
